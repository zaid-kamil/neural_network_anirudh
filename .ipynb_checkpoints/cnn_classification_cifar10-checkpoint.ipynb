```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from tensorflow.keras.datasets import cifar10
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print(x_train.shape)
print(x_test.shape)
```

```python
labels = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
x_train.shape[1:]
```

```python
print(labels[y_train[1,0]])
plt.imshow(x_train[1])
```

```python
plt.figure(figsize=(7,7))
for i in range(1, 26):
    plt.subplot(5,5,i) # 5 rows and 5 cols
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i])
    plt.xlabel(labels[y_train[i][0]])
plt.show()
```

```python
from tensorflow.keras import models, layers
```

```python
model = models.Sequential([
    layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu', input_shape=x_train.shape[1:]),
    layers.MaxPooling2D(),
    layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(units=64, activation='relu'),
    layers.Dense(units=32, activation='relu'),
    layers.Dense(units=10)
])
model.compile(optimizer='adam',
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])
model.summary()
```

```python
tensorboard = tf.keras.callbacks.TensorBoard('logs',histogram_freq=1)
```

```python
x_train.shape
```

```python
history = model.fit(x_train, 
          y_train,
          epochs=5, 
          validation_data=(x_test, y_test),
          callbacks=[tensorboard]
         )
```

```python
def predict_image_from_path(path, size=(32,32)):
    img = Image.open(path)
    inp_img =img.resize(size)
    inp_img = np.array([np.array(inp_img)])
    result = model.predict(inp_img)
    plt.figure(figsize=(2,2))
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    ans = labels[result[0].argmax()]
    return ans, plt.title(ans)
```
